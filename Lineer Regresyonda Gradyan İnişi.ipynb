Doğrusal Regresyon , denetimli öğrenmeye dayalı bir makine öğrenimi algoritmasıdır . Bir regresyon görevi gerçekleştirir . Regresyon, bağımsız değişkenlere dayalı bir hedef tahmin değeri modeller. Çoğunlukla değişkenler ve tahmin arasındaki ilişkiyi bulmak için kullanılır. Farklı regresyon modelleri, dikkate aldıkları bağımlı ve bağımsız değişkenler arasındaki ilişkinin türüne ve kullanılan bağımsız değişkenlerin sayısına göre farklılık gösterir.

Doğrusal regresyon, belirli bir bağımsız değişkene (x) dayalı olarak bir bağımlı değişken değerini (y) tahmin etme görevini gerçekleştirir. Böylece, bu regresyon tekniği x (girdi) ve y(çıktı) arasında doğrusal bir ilişki bulur. Bu nedenle, adı Lineer Regresyondur.
Yukarıdaki şekilde X (girdi) iş tecrübesi ve Y (çıktı) bir kişinin maaşıdır. Regresyon çizgisi, modelimiz için en uygun çizgidir.


Eğim +ve ise : θ j = θ j – (+ve değeri). Dolayısıyla θ j'nin değeri azalır.

Eğim -ve ise : θ j = θ j – (-ve değeri). Dolayısıyla θ j'nin değeri artar.

α'yı çok küçük seçersek, Gradient Descent yerel minimuma ulaşmak için küçük adımlar atacak ve minimuma ulaşmak için daha uzun zaman alacaktır.

Doğrusal Regresyonun Maliyet işlevi(J), tahmin edilen y değeri (öngörülen) ve gerçek y değeri (y) arasındaki Kök Ortalama Karesel Hatadır (RMSE ).

**Gradient Descent :** Maliyet fonksiyonunu azaltmak (RMSE değerini en aza indirerek) ve en uygun çizgiyi elde etmek için θ 1 ve θ 2
değerlerini güncellemek grekir. Buradaki fikir, rastgele θ 1 ve θ 2 değerleriyle başlamak ve ardından değerleri yinelemeli olarak güncelleyerek minimum maliyete ulaşmaktır.
