Makine öğreniminde, gradyan inişi, doğrusal regresyon, lojistik regresyon, sinir ağları vb. gibi algoritmalar için model parametrelerini (katsayılar ve önyargı) hesaplamak için kullanılan bir optimizasyon tekniğidir. Bu teknikte, eğitim seti boyunca tekrar tekrar yinelenir ve modeli güncelleriz. eğitim setine göre hatanın gradyanına göre parametreler. Model parametrelerinin güncellenmesinde dikkate alınan eğitim örneklerinin sayısına bağlı olarak 3 tip gradyan inişimiz vardır:

**Toplu Gradyan İnişi:** Tüm eğitim setine göre hatanın gradyanı hesaplandıktan sonra parametreler güncellenir.

**Stokastik Gradyan İnişi:** Parametreler, tek bir eğitim örneğine göre hatanın gradyanı hesaplandıktan sonra güncellenir.

**Mini-Batch Graadyan İnişi:** Eğitim setinin bir alt kümesine göre hatanın gradyanı hesaplandıktan sonra parametreler güncellenir.

 **Mini-Yığın Gradyan İnişi: Algoritma-**

 ```
 teta = model parametreleri ve max_iters = dönem sayısı olsun. itr = 1, 2, 3, …, max_iters için: mini_batch için (X_mini, y_mini):

X_mini toplu işinde İleri Geçiş:
Mini partide tahminler yapın
Parametrelerin mevcut değerleri ile tahminlerde (J(teta)) hesaplama hatası
Geri Geçiş:
Hesaplama gradyanı(teta) = J(teta) wrt teta'nın kısmi türevi
Parametreleri güncelle:
teta = teta – öğrenme_hızı*gradyan(teta)
```
