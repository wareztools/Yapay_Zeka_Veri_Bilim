**Doğrusal Regresyon:** 

Tahmine dayalı analiz için temel ve yaygın olarak kullanılan türdür. Bağımlı bir değişken ile belirli bir bağımsız değişken kümesi arasındaki ilişkiyi modellemeye yönelik istatistiksel bir yaklaşımdır.

**Bunlar iki tiptir:** 

1. Basit doğrusal Regresyon
2. Çoklu doğrusal gerileme

```
Y = b0 + b1 * x1 + b2 * x2 + b3 * x3 + …… bn * xn 
Y = Bağımlı değişken ve x1, x2, x3, …… xn = çoklu bağımsız değişkenler 
```
**Regresyon Modelinin Varsayımı :**

* Doğrusallık: Bağımlı ve bağımsız değişkenler arasındaki ilişki doğrusal olmalıdır.
* Homoscedasticity: Hataların sabit varyansı korunmalıdır.
* Çok değişkenli normallik: Çoklu Regresyon, artıkların normal dağıldığını varsayar.
* Çoklu Bağlantı Eksikliği: Verilerde çoklu bağlantının çok az olduğu veya hiç olmadığı varsayılır.


**Geçici değişken:**

Çoklu Regresyon Modelinde bildiğimiz gibi birçok kategorik veri kullanıyoruz. Kategorik Verileri kullanmak, sayısal olmayan verileri ilgili Regresyon Modeline dahil etmek için iyi bir yöntemdir. Kategorik Veriler, cinsiyet(erkek/kadın) gibi sabit ve sırasız sayıda değerlerle kategori-veri değerlerini temsil eden veri değerlerini ifade eder. Regresyon modelinde bu değerler Kukla Değişkenler ile temsil edilebilir. 

Bu değişkenler, kategorik değerlerin varlığını ve yokluğunu temsil eden 0 veya 1 gibi değerlerden oluşur.

**Kukla Değişken Tuzağı:**
Kukla Değişken Tuzağı, iki veya daha fazlasının Yüksek Derecede İlişkili olduğu bir durumdur. 

Basit terimle, bir değişkenin diğerinin tahmininden tahmin edilebileceğini söyleyebiliriz. 

Kukla Değişken Tuzağının çözümü, kategorik değişkenlerden birini bırakmaktır. Yani m kukla değişken varsa , modelde m-1 değişkenler kullanılır. 
```
D2 = D1-1   
 Burada D2, D1 = Kukla Değişkenler
 ```


**Model Oluşturma Yöntemi :** 
All-in 
Backward-Elimination
Forward Selection
Bidirectional Elimination
Score Comparison

**Backward-Elimination :**
Adım 1: Modelde başlamak için anlamlı bir seviye seçin. 
Adım #2: Tüm modeli olası tüm tahmin edicilerle sığdırın. 
Adım #3: En yüksek P değerine sahip tahmin ediciyi düşünün. P > SL ise ADIM 4'e gidin, aksi takdirde model Hazırdır. 
Adım #4: Tahmin ediciyi kaldırın. 
Adım #5: Modeli bu değişken olmadan sığdırın.
 

**Forward Selection :**
Adım #1 : Modeli girmek için bir anlamlılık düzeyi seçin (örn. SL = 0.05) 
Adım #2: Tüm basit regresyon modellerini y~ x(n) sığdırın. En düşük P değerine sahip olanı seçin. 
Adım #3: Bu değişkeni saklayın ve zaten sahip olduğunuz modellere ek bir tahminci ekleyerek tüm olası modelleri uygun hale getirin. 
Adım #4: En düşük P değerine sahip tahmin ediciyi düşünün. P < SL ise Adım #3'e gidin, aksi takdirde model Hazırdır.
