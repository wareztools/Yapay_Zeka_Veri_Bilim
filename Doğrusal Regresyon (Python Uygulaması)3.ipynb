#varsayımlar
Aşağıda, bir lineer regresyon modelinin uygulandığı bir veri seti ile ilgili yaptığı temel varsayımlar verilmiştir:  

**Doğrusal ilişki :** Yanıt ve özellik değişkenleri arasındaki ilişki doğrusal olmalıdır. Doğrusallık varsayımı, dağılım grafikleri kullanılarak test edilebilir. Aşağıda gösterildiği gibi, 1. şekil doğrusal olarak ilişkili değişkenleri temsil ederken, 2. ve 3. şekillerdeki değişkenler büyük olasılıkla doğrusal değildir. Bu nedenle, 1. şekil lineer regresyon kullanarak daha iyi tahminler verecektir. 

image.png


**Çok az veya hiç çoklu bağlantı olmaması :** Verilerde çok az veya hiç çoklu bağlantı olmadığı varsayılır. Çoklu doğrusallık, özellikler (veya bağımsız değişkenler) birbirinden bağımsız olmadığında ortaya çıkar.

**Çok az veya hiç otokorelasyon yok:** Diğer bir varsayım, verilerde çok az otokorelasyon olduğu veya hiç olmadığıdır. Otokorelasyon, artık hatalar birbirinden bağımsız olmadığında ortaya çıkar. 


**Homoscedasticity :** Homoscedasticity, hata teriminin (yani, bağımsız değişkenler ile bağımlı değişken arasındaki ilişkideki "gürültü" veya rastgele bozulma) bağımsız değişkenlerin tüm değerlerinde aynı olduğu bir durumu tanımlar. Aşağıda gösterildiği gibi, Şekil 1 eş varyansa sahipken, şekil 2 değişen varyansa sahiptir.

##Uygulamalar:
**Eğilim çizgileri:** Eğilim çizgisi, niceliksel verilerdeki zamanın geçişiyle (GSYİH, petrol fiyatları vb.) değişimi temsil eder. Bu eğilimler genellikle doğrusal bir ilişki izler. Bu nedenle, gelecekteki değerleri tahmin etmek için doğrusal regresyon uygulanabilir. Ancak, diğer olası değişikliklerin verileri etkileyebileceği durumlarda bu yöntem bilimsel geçerlilikten yoksundur.



**Ekonomi:** Doğrusal regresyon, ekonomide baskın ampirik araçtır. Örneğin, tüketici harcamalarını, sabit yatırım harcamalarını, envanter yatırımlarını, bir ülkenin ihracat alımlarını, ithalat harcamalarını, likit varlık tutma talebini, işgücü talebini ve işgücü arzını tahmin etmek için kullanılır.


**Finans:** Sermaye fiyatı varlık modeli, bir yatırımın sistematik risklerini analiz etmek ve ölçmek için doğrusal regresyon kullanır.


**Biyoloji:** Biyolojik sistemlerde parametreler arasındaki nedensel ilişkileri modellemek için doğrusal regresyon kullanılır.
