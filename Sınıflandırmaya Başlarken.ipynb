###Sınıflandırma nasıl çalışır?
 Gerçek sınıflandırma sonuçlarına sahip örnek verileri içeren, eğitim veri seti adı verilen bir dizi gözlemimiz var. Bu veri setinde Classifier adlı bir model eğitiyoruz ve bu modeli belirli bir hastanın hastalığa yakalanıp yakalanmayacağını tahmin etmek için kullanıyoruz. 
Sonuç, şunlara bağlıdır:  

Bu özelliklerin sonucu ne kadar iyi "eşleştirebildiği".

Veri setimizin kalitesi. Nitelik olarak, istatistiksel ve Matematiksel niteliklere atıfta bulunuyorum.

Sınıflandırıcımızın özellikler ve sonuç arasındaki bu ilişkiyi ne kadar iyi genelleştirdiği.

x1 ve x2 değerleri.
