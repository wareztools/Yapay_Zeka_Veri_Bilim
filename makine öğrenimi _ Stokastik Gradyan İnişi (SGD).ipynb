Gradient Descent, çok çeşitli problemlere optimal çözümler bulabilen genel bir optimizasyon algoritmasıdır. 
Genel fikir, maliyet fonksiyonunu en aza indirmek için parametreleri yinelemeli olarak ayarlamaktır.
Gradient Descent'in (GD) 
önemli bir parametresi, öğrenme hızı hiperparametreleri tarafından belirlenen adımların boyutudur. Öğrenme hızı çok küçükse, algoritmanın yakınsaması için birçok iterasyondan geçmesi gerekecek, bu da uzun zaman alacak ve eğer çok yüksekse optimal değeri atlayabiliriz.

Not: Gradient Descent kullanırken, tüm özelliklerin benzer bir ölçeğe sahip olduğundan emin olmalıyız (örn. Scikit-Learn'in StandardScaler sınıfını kullanarak), yoksa yakınsaması çok daha uzun sürer.

##Stokastik Gradyan İnişi (SGD):

Stokastik kelimesi , rastgele bir olasılıkla bağlantılı bir sistem veya süreç anlamına gelir. Bu nedenle, Stokastik Gradyan İniş'te, her yineleme için tüm veri seti yerine rastgele birkaç örnek seçilir. Gradient Descent'de, her yineleme için gradyanı hesaplamak için kullanılan bir veri kümesindeki toplam örnek sayısını belirten "batch" adlı bir terim vardır. Batch Gradient Descent gibi tipik Gradient Descent optimizasyonunda toplu, tüm veri kümesi olarak alınır. Tüm veri setini kullanmak, minimuma daha az gürültülü ve daha az rastgele bir şekilde ulaşmak için gerçekten yararlı olsa da, veri setimiz büyüdüğünde sorun ortaya çıkar. 

Veri kümenizde bir milyon örneğiniz olduğunu varsayalım, bu nedenle, tipik bir Gradient Descent optimizasyon tekniği kullanırsanız, Gradient Descent gerçekleştirirken bir yinelemeyi tamamlamak için bir milyon örneğin tümünü kullanmanız gerekir ve bunun için yapılması gerekir. minimuma ulaşılana kadar her yineleme. Bu nedenle, gerçekleştirilmesi hesaplama açısından çok pahalı hale gelir.

Bu problem Stokastik Gradyan İnişi ile çözülür. SGD'de, her yinelemeyi gerçekleştirmek için yalnızca tek bir numune, yani bir parti boyutu kullanır. Örnek rastgele karıştırılır ve yinelemeyi gerçekleştirmek için seçilir.
