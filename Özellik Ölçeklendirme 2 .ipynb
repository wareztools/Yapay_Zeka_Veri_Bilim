Özellik Ölçekleme, verilerde bulunan bağımsız özellikleri sabit bir aralıkta standartlaştırma tekniğidir. 

Çok değişken büyüklükleri veya değerleri veya birimleri işlemek için veri ön işleme sırasında gerçekleştirilir. 

Özellik ölçeklendirme yapılmazsa, bir makine öğrenimi algoritması, değerlerin biriminden bağımsız olarak daha büyük değerleri, daha yükseği tartma ve daha küçük değerleri daha düşük değerler olarak kabul etme eğilimindedir.

**Örnek:** Bir algoritma öznitelik ölçekleme yöntemini kullanmıyorsa, 3000 metre değerini 5 km'den büyük kabul edebilir ancak bu aslında doğru değildir ve bu durumda algoritma yanlış tahminler verecektir. Bu nedenle, tüm değerleri aynı büyüklüklere getirmek için Özellik Ölçeklendirmeyi kullanıyoruz ve böylece bu sorunu çözüş oluyoruz.

##Özellik Ölçekleme gerçekleştirme teknikleri
En önemli iki tanesini göz önünde bulunduralım:

**Min-Maks Normalleştirme:** 
Bu teknik, bir özelliği veya gözlem değerini 0 ile 1 arasında dağılım değeriyle yeniden ölçeklendirir.

image.png

**Standardizasyon:** Bir öznitelik değerini, 0 ortalama değeri ve varyansı 1 olacak şekilde yeniden ölçeklendiren çok etkili bir tekniktir.

image.png



##Örnekler


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn import preprocessing
```

```python
data = pd.read_csv("/content/Data_for_Missing_Values.csv")
data.head()

x = data.iloc[:,1:3].values
print("\n Orjinal data görüntüsü :\n",x)
```

Eksik Değerlerin İşlenmesi


```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0,1))

x_after_min_max_scaler = min_max_scaler.fit_transform(x)

print('\n Min max Scaler sonrası :\n',x_after_min_max_scaler)
```

```python
standardisation = preprocessing.StandardScaler()

x_after_Standardisation = Standardisation.fit_transform(x)

print("\n Standardizasyon sonrası :\n",x_after_Standardisation)
```
