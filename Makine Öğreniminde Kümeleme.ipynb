##DBSCAN: Gürültülü Uygulamaların Yoğunluk Tabanlı Uzamsal Kümelenmesi 
Bu veri noktaları, veri noktasının küme merkezinden verilen kısıtlama içinde yer aldığı temel kavramı kullanılarak kümelenir. Aykırı değerlerin hesaplanmasında çeşitli uzaklık yöntem ve teknikleri kullanılmaktadır. 

###Neden Kümeleme?  
Kümeleme, mevcut etiketlenmemiş veriler arasındaki içsel gruplamayı belirlediği için çok önemlidir. İyi kümeleme için hiçbir kriter yoktur. Kullanıcıya bağlıdır, ihtiyaçlarını karşılayan kriterleri kullanabilirler. Örneğin, homojen gruplar için temsilciler bulmak (veri azaltma), “doğal kümeler” bulmak ve bunların bilinmeyen özelliklerini tanımlamak (“doğal” veri türleri), faydalı ve uygun gruplamalar (“yararlı” veri sınıfları) bulmakla ilgilenebiliriz. veya olağandışı veri nesnelerini bulmada (outlier tespiti). Bu algoritma, noktaların benzerliğini oluşturan bazı varsayımlarda bulunmalı ve her varsayım farklı ve eşit derecede geçerli kümeler oluşturmalıdır. 

###Kümeleme Yöntemleri : 

**Yoğunluğa Dayalı Yöntemler:** Bu yöntemler, kümeleri, uzayın alt yoğun bölgesinden bazı benzerlik ve farklılıklara sahip yoğun bölge olarak kabul eder. Bu yöntemler iyi bir doğruluğa ve iki kümeyi birleştirme yeteneğine sahiptir. Örnek DBSCAN (Gürültü ile Uygulamaların Yoğunluk Tabanlı Mekansal Kümelenmesi) , OPTİK (Kümeleme Yapısını Tanımlamak için Sıralama Noktaları) vb.


**Hiyerarşik Tabanlı Yöntemler:** Bu yöntemde oluşturulan kümeler, hiyerarşiye dayalı ağaç tipi bir yapı oluşturur. Daha önce oluşturulan kümeler kullanılarak yeni kümeler oluşturulur. İki kategoriye ayrılır 
Aglomeratif (aşağıdan yukarıya yaklaşım )
Bölücü (yukarıdan aşağıya yaklaşım 

**Bölümleme Yöntemleri:** Bu yöntemler nesneleri k kümeye böler ve her bölüm bir küme oluşturur. Bu yöntem, örneğin , K-ortalamalar, CLARANS (Rastgele Aramaya Dayalı Büyük Uygulamaları Kümeleme) vb. gibi mesafenin ana parametre olduğu durumlar gibi bir nesnel kriter benzerlik fonksiyonunu optimize etmek için kullanılır.

**Izgara Tabanlı Yöntemler:** Bu yöntemde, veri alanı, ızgara benzeri bir yapı oluşturan sonlu sayıda hücreye formüle edilir. Bu gridler üzerinde yapılan tüm kümeleme işlemleri hızlıdır ve veri nesnesi sayısından bağımsızdır, örneğin STING (Statistical Information Grid), wave cluster, CLIQUE (Clustering In Quest) vb.

**Kümelemenin farklı alanlardaki uygulamaları** 

* Pazarlama: Pazarlama amaçları için müşteri segmentlerini karakterize etmek ve keşfetmek için kullanılabilir.
* Biyoloji: Farklı bitki ve hayvan türleri arasında sınıflandırma yapmak için kullanılabilir.
* Kütüphaneler: Farklı kitapların konu ve bilgi bazında kümelenmesinde kullanılır.
* Sigorta: Müşterileri, poliçelerini onaylamak ve dolandırıcılıkları tespit etmek için kullanılır.
