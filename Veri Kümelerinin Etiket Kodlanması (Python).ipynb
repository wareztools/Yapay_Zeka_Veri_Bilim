Makine öğreniminde genellikle bir veya birden fazla sütunda birden çok etiket içeren veri kümeleriyle ilgileniriz. Bu etiketler kelime veya sayı şeklinde olabilir. Verileri anlaşılabilir veya insan tarafından okunabilir hale getirmek için eğitim verileri genellikle kelimelerle etiketlenir. 

**Etiket Kodlama** , etiketlerin makine tarafından okunabilir biçime dönüştürülmesi için sayısal bir biçime dönüştürülmesi anlamına gelir. 

Makine öğrenimi algoritmaları daha sonra bu etiketlerin nasıl çalıştırılması gerektiğine daha iyi karar verebilir. Denetimli öğrenmede yapılandırılmış veri seti için önemli bir ön işleme adımıdır.

**Örnek**
Diyelim ki bazı veri kümelerinde bir Boyut sütunumuz var .

Burada 0, uzun için etiket, 1 orta için etiket ve 2, kısa yükseklik için bir etikettir.


Species olan hedef sütundaki iris veri kümesine Label Encoding uyguluyoruz.
 Üç tür Iris-setosa, Iris-versicolor, Iris-virginica içerir . 

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/iris.csv')
df['variety'].unique()
```

```python
import numpy as np
import pandas as pd 
from sklearn import preprocessing

label = preprocessing.LabelEncoder()

df['variety'] = label.fit_transform(df['variety'])

df['variety'].unique()
```

**Etiket kodlamasının sınırlandırılması** 

Etiket kodlaması, verileri makine tarafından okunabilir biçimde dönüştürür, ancak her veri sınıfına benzersiz bir sayı (0'dan başlayarak) atar. Bu, veri setlerinin eğitiminde öncelikli konuların oluşmasına neden olabilir. 

Yüksek bir değere sahip bir etiket, daha düşük bir değere sahip bir etiketten daha yüksek önceliğe sahip olarak kabul edilebilir.

**Örnek** Meksika, Paris, Dubai
çıktı sınıflarına sahip bir nitelik . Label Encoding'de bu sütun Meksika'nın 0 , Paris'in 1 , Dubai'nin 2 ile değiştirilmesini  sağlar .

 Bununla Dubai'nin modeli eğitirken Meksika ve Paris'ten daha yüksek önceliğe sahip olduğu yorumlanabilir , Ama aslında, burada bu şehirler arasında böyle bir öncelik ilişkisi yoktur.
