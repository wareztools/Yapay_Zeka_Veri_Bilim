##Regresyon Ne Zaman Seçilir  
Bir regresyon problemi, çıktı değişkeninin “maaş” veya “ağırlık” gibi gerçek veya sürekli bir değer olması durumudur. Birçok farklı model kullanılabilir, en basiti lineer regresyondur. Noktalardan geçen en iyi hiperdüzlem ile verileri sığdırmaya çalışır.



**Regresyon Analizi** , bağımlı değişkenler veya kriter değişkenler ile bir veya daha fazla bağımsız değişken veya tahmin edici arasındaki ilişkileri tahmin etmeye yönelik istatistiksel bir süreçtir. Regresyon analizi, seçilen tahmin edicilerdeki değişikliklerle ilgili olarak kriterlerdeki değişiklikleri açıklar. Kriterlerin koşullu beklentisi, bağımsız değişkenler değiştirildiğinde bağımlı değişkenlerin ortalama değerinin verildiği yordayıcılara dayanmaktadır. Regresyon analizi için üç ana kullanım, tahmin edicilerin gücünü belirlemek, bir etkiyi tahmin etmek ve trend tahminidir

##Regresyon Türleri:  
1. Doğrusal Regresyon **Y' = bX + A.**

 Doğrusal regresyon, kriter veya skaler yanıt ile çoklu tahmin ediciler veya açıklayıcı değişkenler arasındaki ilişkiyi modellemek için doğrusal bir yaklaşımdır. 

 Doğrusal regresyon, tahmin edicilerin değerleri verilen yanıtın koşullu olasılık dağılımına odaklanır. 

 Doğrusal regresyon için, fazla uydurma tehlikesi vardır

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
x = 11*np.random.random((10,1))

y = 1.0*x+3.0

model = LinearRegression()
model.fit(x,y)

x_tahmin = np.linspace(0,11,100)
y_tahmin = model.predict(x_tahmin[:,np.newaxis])

plt.figure(figsize = (3,5))
ax = plt.axes()
ax.scatter(x,y)


ax.plot(x_tahmin,y_tahmin)
ax.set_xlabel('Tahmin Edici')
ax.set_ylabel('kriter')
ax.axis('tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = 11 * np.random.random((10, 1))

# y = a * x + b
y = 1.0 * x + 3.0

# doğrusal bir regresyon modeli oluşturun
model = LinearRegression()
model.fit(x, y)

# x'in x'ten tahmin edildiği verilerden y'yi tahmin edin
x_pred = np.linspace(0, 11, 100)
y_pred = model.predict(x_pred[:, np.newaxis])

# sonuçları çiz
plt.figure(figsize =(3, 5))
ax = plt.axes()
ax.scatter(x, y)

ax.plot(x_pred, y_pred)
ax.set_xlabel(' tahmin ediciler ')
ax.set_ylabel(' kriter ')
ax.axis(' sıkı ')

plt.show()
```
