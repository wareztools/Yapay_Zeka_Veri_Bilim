##Diziler

Numpy'deki dizi, bir grup pozitif tamsayı tarafından indekslenen, hepsi aynı tipte bir elementler (genellikle sayılar) tablosudur. Numpy'de dizinin boyut sayısı, dizinin rankı olarak adlandırılır. Her boyut boyunca dizinin boyutunu veren bir tamsayı demeti, dizinin şekli olarak bilinir. Numpy'deki bir dizi sınıfı ndarray olarak adlandırılır . Numpy dizilerindeki öğelere köşeli parantezler kullanılarak erişilir ve iç içe Python Listeleri kullanılarak başlatılabilir.


###Bir Numpy Kütüphanesinde Numpy Dizisi Oluşturma 
Diziler, Dizinin boyutunu tanımlayan çeşitli Sıra sayılarıyla birden çok yolla oluşturulabilir. Diziler, listeler, demetler, vb. gibi çeşitli veri türlerinin kullanımıyla da oluşturulabilir. Elde edilen dizinin türü, dizilerdeki öğelerin türünden çıkarılır.
Not: Dizi türü, dizi oluşturulurken açıkça tanımlanabilir.

```python
import numpy as np

dizi = np.array([1, 2, 3])
print("1.dereceden dizi \n",dizi )

dizi = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("2.dereceden dizi \n",dizi)

dizi = np.array((1, 5, 9))
print("\n tuple tipinde dizi \n",dizi)
```

###Dizi Dizine Erişme
Numpy dizisinde dizin oluşturma veya dizi dizinine erişim birden çok yolla yapılabilir. Bir dizi aralığını yazdırmak için dilimleme yapılır. Bir dizinin dilimlenmesi, orijinal diziden bir dizi öğeyi yazdırmak için kullanılan yeni bir dizide bir aralığı tanımlamaktır. Dilimlenmiş dizi, orijinal dizinin bir dizi öğesini içerdiğinden, içeriği dilimlenmiş dizi yardımıyla değiştirmek orijinal dizi içeriğini değiştirir.

```python
import numpy as np

dizi = np.array([[-1, 8, 5, 6],
                  [-9, 7, 2, 1],
                    [-5, -3, 8.7, 1.3],
                      [-7, 14, 9.41, 0 ]])
print("İlk dizi \n", dizi)

dilimlenmiş_dizi = dizi[:2 , ::3]
print("dilimleniş çıktı (ilk 2 satır)\n",dilimlenmiş_dizi)

dizi_içi = dizi[[1, 1, 0, 3],
                [3, 2, 0, 1]]
print("dizi içi\n",dizi_içi)
```

###Temel Dizi İşlemleri
Numpy'de diziler, belirli bir dizide veya Dizilerin bir kombinasyonunda gerçekleştirilebilecek çok çeşitli işlemlere izin verir. Bu işlemler, bazı temel Matematiksel işlemlerin yanı sıra Tekli ve İkili işlemleri içerir.

```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
 
b = np.array([[4, 3],
               [2, 1]])
print("A dizisinin her element  1 ekle :\n", a + 1 )
print("B dizisinin her elementinden 2 çıkar :\n", b - 2 )
print("A dizisin elementlerinin toplamı :\n", a.sum())
print("A dizisi ile B dizisinin elementlerini topla:\n", a + b )
```

##Numpy'deki Veri Türleri
Her Numpy dizisi, bir grup pozitif tamsayı tarafından indekslenen, tümü aynı tipte bir elementler tablosudur (genellikle sayılar). Her ndarray'in ilişkili bir veri türü (dtype) nesnesi vardır. Bu veri türü nesnesi (dtype), dizinin düzeni hakkında bilgi sağlar. Bir ndarray'in değerleri, dtype nesnesi tarafından yorumlanabilen bitişik bir bellek bayt bloğu olarak düşünülebilecek bir arabellekte saklanır. Numpy, diziler oluşturmak için kullanılabilecek çok sayıda sayısal veri türü sağlar. Dizi oluşturma sırasında, Numpy bir veri türünü tahmin etmeye çalışır, ancak dizileri oluşturan işlevler genellikle veri türünü açıkça belirtmek için isteğe bağlı bir argüman da içerir.

###Bir Veri Türü Nesnesi
Numpy'de, belirli bir veri türü gerekmedikçe Dizilerin veri türlerinin tanımlanması gerekmez. Numpy, yapıcı işlevinde önceden tanımlanmamış Diziler için veri türünü tahmin etmeye çalışır.

```python
import numpy as np
x = np.array([4,9])
print(x,"dizi tipi",x.dtype)

x = np.array([4.8, 9.0])
print(x,"dizi tipi",x.dtype )

x = np.array([1,2], dtype = np.float64)
print(x.dtype)
```

###DataType dizisinde Matematik İşlemleri
Numpy dizilerinde, temel matematiksel işlemler dizi üzerinde eleman bazında gerçekleştirilir. Bu işlemler hem operatör aşırı yüklemeleri hem de işlevler olarak uygulanır. Numpy'de , dizi öğelerinin eklenmesi için sum :, öğelerin devrik için T : için vb.

```python
import numpy as np

dizi1 = np.array([[4, 7], [2,6]],
                 dtype = np.float64)

dizi2 = np.array([[3,8],[8,9]],
                 dtype = np.float64)

sum = np.add(dizi1, dizi2)
print("2 dizinin toplamı :\n",sum)

sum1 = np.sum(dizi1)
print("dizi 1 elementlerinin toplamı :\n",sum1)

sqrt = np.sqrt(dizi1)
print("Dizi 1 elementlerinin farkı :\n",sqrt)

trans_arr = dizi1.T
print("dizi 1 trans :\n",trans_arr) 
```

##Eksik Verilerle Çalışmak
Eksik Veriler, bir veya daha fazla öğe veya tüm birim için hiçbir bilgi sağlanmadığında ortaya çıkabilir. Eksik Veri, gerçek hayat senaryosunda çok büyük bir sorundur. Eksik Veri, pandalarda NA (Kullanılamaz) değerleri olarak da ifade edilebilir.

###Eksik Değer Kontrolü
 Pandas DataFrame'deki eksik değerleri kontrol etmek için isnull()ve notnull()  işlevini kullanırız . Her iki işlev de bir değerin olup olmadığını kontrol etmeye yardımcı olur. 
 NaN. Bu işlev, bir dizide boş değerler bulmak için Pandas Series'de de kullanılabilir.

```python
import pandas as pd
import numpy as np

veri = { 'İlk Sayı'  : [100, 80, np.nan, 75],
        'İkinci Sayı' : [50, 64, 74, np.nan],
        'Üçüncü Sayı' : [np.nan, np.nan, 52, 78]}  

df = pd.DataFrame(veri)
df.isnull()
```

###Eksik Değerleri Doldurmak
Bir veri kümesindeki boş değerleri doldurmak için fillna() kullanırız ve bu işlev  NaN değerlerini kendi değerleriyle değiştirir. Tüm bu işlevler, bir DataFrame'in veri kümelerinde boş değerlerin doldurulmasına yardımcı olur. Interpolate() işlevi temel olarak dataframesindeki değerleri doldurmak için kullanılır, ancak değeri sabit kodlamak yerine eksik değerleri doldurmak için çeşitli enterpolasyon teknikleri kullanır.replace()interpolate()NA

```python
import pandas as pd
import numpy as np

veri = { 'İlk Sayı'  : [100, 80, np.nan, 75],
        'İkinci Sayı' : [50, 64, 74, np.nan],
        'Üçüncü Sayı' : [np.nan, np.nan, 52, 78]}  
df = pd.DataFrame(veri)
df.fillna(10)
```

##Satırlar ve Sütünlar Üzeinde Yineleme

Yineleme, bir şeyin her bir öğesini birbiri ardına almak için genel bir terimdir. Pandas DataFrame satırlar ve sütunlardan oluşur, bu nedenle dataframesi üzerinde yineleme yapmak için bir dataframesini sözlük gibi yinelememiz gerekir.



###Satırlar Üzerinde Yineleme
yineleme yapmak için üç işlev kullanabiliriz iteritems(), iterrows(), itertuples() . 
Bu üç işlev, satırlar üzerinde yineleme yapmamıza  yardımcı olacaktır.

```python
import pandas as pd
import numpy as np

veri = { 'İsim'  : ['Avery', 'jae', 'jonas', 'jordan'],
        'Yaş' : [25, 27, 32, 23],
        'Takım' : ['Boston', 'Boston', 'Boston', 'Boston']}  
df = pd.DataFrame(veri)
df
```

her bir satır öğesini elde etmek için iterrows() işlevi uyguluyoruz.

```python
import pandas as pd
import numpy as np

veri = { 'İsim'  : ['Avery', 'jae', 'jonas', 'jordan'],
        'Yaş' : [25, 27, 32, 23],
        'Takım' : ['Boston', 'Boston', 'Boston', 'Boston']}  
df = pd.DataFrame(veri)

for i, j in df.iterrows():
  print(i ,j)
  print()
```
