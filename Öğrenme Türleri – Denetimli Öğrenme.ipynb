##Denetimli Öğrenme
Denetimli öğrenme , modelin etiketli bir veri kümesi üzerinde eğitilmesidir. Etiketli bir veri kümesi, hem giriş hem de çıkış parametrelerine sahip olan bir veri kümesidir . Hem eğitim hem de doğrulama bu tür öğrenmede, veri kümeleri aşağıdaki şekillerde gösterildiği gibi etiketlenir. 

image.png

>**Şekil A:** Bir müşterinin söz konusu ürünü cinsiyetine, yaşına ve maaşına göre satın alıp almayacağını tahmin etmede faydalı olan bir alışveriş mağazasının veri setidir.

**Girdi:** Cinsiyet, Yaş, Maaş 

**Çıktı:** Satın alındı, yani 0 veya 1; 1, evet, müşterinin satın alacağı anlamına gelir ve 0, müşterinin satın almayacağı anlamına gelir. 
>**Şekil B:** Rüzgar hızını farklı parametrelere göre tahmin etme amacına hizmet eden bir Meteorolojik veri setidir. 

**Girdi:** Çiy Noktası, Sıcaklık, Basınç, Bağıl Nem, Rüzgar Yönü 

**Çıkış:**Rüzgar Hızı 


**Sistemi eğitmek:** 
Modeli eğitirken, veriler genellikle 80:20 oranında bölünür yani %80 eğitim verisi, kalanı test verisi olarak. Eğitim verilerinde, verilerin %80'i için girdinin yanı sıra çıktıyı da besliyoruz. Model yalnızca eğitim verilerinden öğrenir.

 Modelimizi oluşturmak için (sonraki makalelerde ayrıntılı olarak tartışacağımız) farklı makine öğrenme algoritmaları kullanıyoruz. Öğrenme, modelin kendi başına bir mantık inşa edeceği anlamına gelir. 

 
Model hazır olduğunda test edilmesi iyidir. Test sırasında, girdi, modelin daha önce hiç görmediği kalan %20'lik veriden beslenir, model bir değer tahmin eder ve biz bunu gerçek çıktıyla karşılaştırır ve doğruluğu hesaplarız.

###Denetimli Öğrenme Türleri
1. Sınıflandırma 

Çıktının tanımlanmış etiketlere (ayrık değer) sahip olduğu bir Denetimli Öğrenme görevidir. 

Örneğin, yukarıdaki Şekil A'da, Çıktı – Satın Alınan, 0 veya 1 gibi tanımlanmış etiketlere sahiptir; 1, müşterinin satın alacağı anlamına gelir ve 0, müşterinin satın almayacağı anlamına gelir. Buradaki amaç, belirli bir sınıfa ait ayrık değerleri tahmin etmek ve doğruluk temelinde değerlendirmektir. 
İkili veya çok sınıflı sınıflandırma olabilir. İkili sınıflandırmada , model ya 0 ya da 1 öngörür; evet veya hayır, ancak çok sınıflı sınıflandırma durumunda, model birden fazla sınıfı tahmin eder.

Örnek: Gmail, postaları sosyal, promosyonlar, güncellemeler ve forumlar gibi birden fazla sınıfta sınıflandırır.

2. Regresyon
 Çıktının sürekli değere sahip olduğu bir Denetimli Öğrenme görevidir.

Örneğin, yukarıdaki Şekil B'de, Çıkış – Rüzgar Hızı herhangi bir ayrık değere sahip değildir ancak belirli bir aralıkta süreklidir. Buradaki amaç, modelimizin yapabileceği kadar gerçek çıktı değerine daha yakın bir değer tahmin etmektir ve ardından hata değeri hesaplanarak değerlendirme yapılır. Hata ne kadar küçükse, regresyon modelimizin doğruluğu o kadar yüksek olur.

**Denetimli Öğrenme Algoritmaları Örneği:**  

* Doğrusal Regresyon
* Lojistik regresyon
* En yakın komşu
* Gauss Saf Bayes
* Karar ağaçları
* Destek Vektör Makinesi (SVM)
* Random Forest
