 dronekit-python deposunu Github'dan klonlayalım
```
git clone http://github.com/dronekit/dronekit-python.git
```
Çalıştırmak istediğimiz örneğe gidelim (veya sonraki adımda tam yolu belirtin). Örneklerin tümü dronekit-python/examples/ alt dizinlerinde saklanır .
```
cd dronekit-python/examples/vehicle_state/
```
Komut dosyası tarafından başlatılan/yönetilen bir simülatöre bağlanmak için:
```
python vehicle_state.py
````
Belirli bir araca bağlanmak için, bağlantı dizesini bağımsız değişken aracılığıyla connect kullanılır
```
python vehicle_state.py --connect udpin:0.0.0.0:14550
```
