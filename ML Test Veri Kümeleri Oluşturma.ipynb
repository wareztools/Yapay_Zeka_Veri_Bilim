##Normal Dağılım
Olasılık teorisinde, normal veya Gauss dağılımı, ortalamaya yakın verilerin, ortalamadan uzak verilerden daha sık meydana geldiğini gösteren, ortalamaya göre simetrik olan çok yaygın bir sürekli olasılık dağılımıdır. 

İstatistikte kullanılan normal dağılımlar ve genellikle gerçek değerli rastgele değişkenleri temsil etmek için kullanılır. 
Normal dağılım, istatistiksel analizlerde en yaygın dağılım türüdür.

 Standart normal dağılımın iki parametresi vardır: ortalama ve standart sapma. Ortalama, dağılımın merkezi eğilimidir. Standart sapma, bir değişkenlik ölçüsüdür. 
 
Normal dağılımın genişliğini tanımlar. Standart sapma, değerlerin ortalamadan ne kadar uzağa düşme eğiliminde olduğunu belirler.

 Gözlemler ve ortalama arasındaki tipik mesafeyi temsil eder. Birçok doğal fenomene uyar, örneğin boylar, kan basıncı, ölçüm hatası ve IQ puanları normal dağılımı takip eder.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
ort = 0.5
sigma = 0.1

np.random.seed(0)

X = np.random.normal(ort,sigma, (495,1))
Y = np.random.normal(ort*2,sigma*3, (495,1))

plt.scatter(X,Y,color = 'r')
plt.show()
```
