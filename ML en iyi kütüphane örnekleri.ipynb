Scikit-learn, klasik ML algoritmaları için en popüler ML kitaplıklarından biridir. NumPy ve SciPy olmak üzere iki temel Python kütüphanesinin üzerine inşa edilmiştir. Scikit-learn, denetimli ve denetimsiz öğrenme algoritmalarının çoğunu destekler. Scikit-learn aynı zamanda veri madenciliği ve veri analizi için de kullanılabilir, bu da onu makine öğrenimi ile başlayanlar için harika bir araç haline getirir. 

````python
# Scikit-learn kullanan Python betiği
# Karar Ağacı Sınıflandırıcısı için
  
# Örnek Karar Ağacı Sınıflandırıcısı
from sklearn import datasets
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier

# iris veri kümelerini yükle
dataset = datasets.load_iris()

# verilere bir CART modeli sığdır
model = DecisionTreeClassifier()
model.fit(dataset.data, dataset.target)
print(model)

# tahminlerde bulunmak
expected = dataset.target
predicted = model.predict(dataset.data)

# modelin uyumunu özetlemek
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
````

PyTorch, Lua'da bir sarmalayıcı ile C'de uygulanan açık kaynaklı bir Makine Öğrenimi kitaplığı olan Torch'a dayalı Python için popüler bir açık kaynaklı Makine Öğrenimi kitaplığıdır. Computer Vision, Natural Language Processing (NLP) ve daha birçok ML programını destekleyen kapsamlı bir araç ve kitaplık seçeneklerine sahiptir. Geliştiricilerin GPU hızlandırmalı Tensörler üzerinde hesaplamalar yapmasına olanak tanır ve ayrıca hesaplamalı grafikler oluşturmaya yardımcı olur. 
 

````python
# PyTorch kullanan Python programı
# uygun tensörleri tanımlamak için
# iki katmanlı ağdan rastgele
# veri ve kaybın hesaplanması

import torch


dtype = torch.float
device = torch.device("cpu")
device = torch.device("cuda:0") #GPU'da çalıştırmak için bunun yorumunu kaldır

# N, parti boyutudur; D_in girdi boyutudur;
# H gizli boyuttur; D_out çıktı boyutudur.
N, D_in, H, D_out = 64, 1000, 100, 10

# Rastgele giriş ve çıkış verileri oluşturun
x = torch.random(N, D_in, device = device, dtype = dtype)
y = torch.random(N, D_out, device = device, dtype = dtype)

# Ağırlıkları rastgele başlat
w1 = torch.random(D_in, H, device = device, dtype = dtype)
w2 = torch.random(H, D_out, device = device, dtype = dtype)

learning_rate = 1e-6
for t in range(500):
	# İleri geçiş: tahmin edilen y'yi hesapla
	h = x.mm(w1)
	h_relu = h.clamp(min = 0)
	y_pred = h_relu.mm(w2)

	# Hesaplama ve yazdırma kaybı
	loss = (y_pred - y).pow(2).sum().item()
	print(t, loss)

  # Kayba göre w1 ve w2 gradyanlarını hesaplamak için Backprop
	grad_y_pred = 2.0 * (y_pred - y)
	grad_w2 = h_relu.t().mm(grad_y_pred)
	grad_h_relu = grad_y_pred.mm(w2.t())
	grad_h = grad_h_relu.clone()
	grad_h[h < 0] = 0
	grad_w1 = x.t().mm(grad_h)

	# Gradyan inişini kullanarak ağırlıkları güncelleyin
	w1 -= learning_rate * grad_w1
	w2 -= learning_rate * grad_w2
````
