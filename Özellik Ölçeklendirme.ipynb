#Özellik Ölçekleme
Verilerde bulunan bağımsız özellikleri sabit bir aralıkta standartlaştırma tekniğidir. 

Veri ön işleme sırasında gerçekleştirilir. 
#Çalışma
Her biri bu bağımsız veri özelliklerine sahip olan 5000 kişilik veri boyutuna sahip  Yaş , Maaş , Apartman
özelliklerine sahip bir veri seti verildi.

**Her veri noktası şu şekilde etiketlenir:**
* **Sınıf1- EVET** (Belirtilen Yaş , Maaş , Apartman özellik değeri ile satın alabileceğiniz gayrimenkulü ifade eder)
* **Sınıf2- HAYIR** (Belirtilen Yaş , Maaş ,  Apartman özellik değeri ile mülkün satın alınamayacağı anlamına gelir).


*Modeli eğitmek için bir veri seti kullanarak, verilen özellik değerleri ile bir mülkün satın alınıp alınamayacağını tahmin edebilen bir model inşa etmeyi amaçlar.*

Model eğitildikten sonra, verilen veri kümesinden veri noktaları ile N-boyutlu (burada N, veri kümesinde bulunan özelliklerin sayısıdır) grafik oluşturulabilir. Aşağıda verilen şekil, modelin ideal bir temsilidir.
  
image.png

Şekilde görüldüğü gibi yıldız veri noktaları **Sınıf1 – Evet**’e aittir ve daireler **Sınıf2 – Hayır** etiketlerini temsil eder ve model bu veri noktaları kullanılarak eğitilir.
 
Şimdi yeni bir veri noktası (şekilde gösterildiği gibi elmas) verilmiştir ve yukarıda belirtilen 3 özellik ( Yaş , Maaş , Apartman ) için farklı bağımsız değerlere sahiptir. Model, bu veri noktasının Evet veya Hayır'a ait olup olmadığını tahmin etmelidir.

**Yeni veri noktalarının sınıfının tahmini:** 
Model, bu veri noktasının her sınıf grubunun merkezinden uzaklığını hesaplar. Son olarak, bu veri noktası, ondan minimum bir centroid mesafesine sahip olacak olan o sınıfa ait olacaktır. 

Aşşağıda ki yöntemler kullanılarak merkez ve veri noktası arasındaki mesafe hesaplanabilir. 

**Öklid Mesafesi:** Her sınıfın veri noktası ve ağırlık merkezinin koordinatları (özellik değerleri – Yaş , Maaş ,  Apartman ) arasındaki farkların karelerinin toplamının kare köküdür . 

Bu formül Pisagor teoreminden  verilir.

burada öznitelik değerleri, x, Veri Noktası değeridir, y, Centroid değeridir ve k, hayırdır.  

Örnek: verilen veri kümesinde k = 3 vardır

**Manhattan Mesafesi:** Her sınıfın veri noktası ve ağırlık merkezinin koordinatları (özellik değerleri) arasındaki mutlak farkların toplamı olarak hesaplanır. 

image.png 
 
**Minkowski Mesafesi:** Yukarıdaki iki yöntemin bir genellemesidir.

##Özellik Ölçeklendirme İhtiyacı: 
Verilen veri seti 3 özellik içermektedir – Yaş , Maaş ,  Apartman . Yaş için 10-60 , Maaş için 1 LAC  - 40 LAC , Apartman  için 1 - 5 aralığını düşünün . Tüm bu özellikler birbirinden bağımsızdır. 

Sınıf 1'in ağırlık merkezinin [40, 22 LAC, 3] ve tahmin edilecek veri noktasının [57, 33 LAC, 2] olduğunu varsayalım. 

1 LAC =  0,0194 dolar

**Örnek**

Mesafe = (|(40 - 57)| + |(2200000 - 3300000)| + |(3 - 2)|)

Verilen veri noktasının sınıfını tahmin ederken Maaş özelliğinin diğer tüm özelliklere hakim olacağı ve tüm özellikler birbirinden bağımsız olduğu için yani kişinin maaşının yaşıyla veya dairenin hangi ihtiyacı ile ilgisi olmadığı görülmektedir. Bu, modelin her zaman yanlış tahmin edeceği anlamına gelir. 

Dolayısıyla, bu sorunun basit çözümü Özellik Ölçeklendirmedir. Özellik Ölçekleme Algoritmaları, Yaş, Maaş, Apartmanı [-1, 1] veya [0, 1] gibi sabit bir aralıkta ölçeklendirir. Ve sonra hiçbir özellik diğerlerine hükmedemez.
