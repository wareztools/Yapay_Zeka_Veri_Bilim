##Bayes teoremi

Bayes Teoremi, daha önce gerçekleşmiş başka bir olayın olasılığı verildiğinde, bir olayın olma olasılığını bulur. Bayes teoremi matematiksel olarak aşağıdaki denklemle ifade edilir:

image.png

burada A ve B olaylardır ve P(B) ≠ 0.

* Temel olarak, B olayının doğru olduğu göz önüne alındığında, A olayının olasılığını bulmaya çalışıyoruz. Olay B ayrıca kanıt olarak adlandırılır .
* P(A), A'nın a prioridir (önceki olasılık, yani kanıt görülmeden önce olayın olasılığı). Kanıt, bilinmeyen bir örneğin öznitelik değeridir (burada, olay B'dir).
* P(A|B), B'nin posteriori olasılığıdır, yani kanıt görüldükten sonra olayın olasılığı.
