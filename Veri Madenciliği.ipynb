	
**Veri Madenciliği :** 

Genel anlamda veri madenciliği, örüntüler elde etmek ve bu örüntü hakkında bilgi edinmek için farklı formlardaki verilerin madenciliği veya derinlerine inilmesi anlamına gelir. 

Veri madenciliği sürecinde önce büyük veri kümeleri sıralanır, ardından örüntüler belirlenir ve veri analizi yapmak ve sorunları çözmek için ilişkiler kurulur. 

**Sınıflandırma :** Bir veri analizi görevidir, yani veri sınıflarını ve kavramlarını tanımlayan ve ayırt eden bir model bulma sürecidir. Sınıflandırma, gözlemleri içeren ve kategori üyeliği bilinen bir eğitim veri setine dayanarak, yeni bir gözlemin bir dizi kategoriden (alt popülasyonlar) hangisine ait olduğunu belirleme problemidir.

**Eğitim ve Test:** 
Diyelim ki bir vantilatörün altında oturan biri var ve vantilatör üzerine düşmeye başlıyor, yaralanmamak için kenara çekilmesi gerekiyor. Yani, bu onun uzaklaşması için eğitim kısmı. Test yapılırken, kişinin kendisine doğru gelen veya üzerine düşen ağır bir cisim görüp kenara çekilip çekilmediği test edilirken sistem olumlu, kişi yana çekilmiyorsa sistem olumsuz olarak test edilir. 
Aynı durum verilerde de geçerlidir, doğru ve en iyi sonuçların alınabilmesi için eğitilmelidir.


Veri madenciliğiyle ilişkili, bize dosyanın biçimini (metin biçiminde mi yoksa sayısal biçimde mi olduğunu) söyleyen belirli veri türleri vardır. 
Nitelikler – Bir nesnenin farklı özelliklerini temsil eder. Farklı nitelik türleri şunlardır: 

##**Binary :** 

>Yalnızca iki değere sahiptir, yani Doğru veya Yanlış 
Örnek: Diyelim ki bazı ürünleri değerlendiren bir anket var. Yararlı olup olmadığını kontrol etmemiz gerekiyor. Bu nedenle Müşteri, Evet veya Hayır olarak yanıtlamalıdır. 
Ürün kullanışlılığı: Evet / Hayır

>>**Simetrik :**  Her iki değer de her açıdan eşit derecede önemlidir

>>**Asimetrik :**  Her iki değer de önemli olmayabilir.


##**Nominal :** 
>İkiden fazla sonuç mümkün olduğunda. Tamsayı biçiminde olmaktan ziyade Alfabe biçimindedir.

Örnek : Bir malzeme seçmelisiniz, ancak farklı renklerde. Yani renk Sarı, Yeşil, Siyah, Kırmızı olabilir. 

Farklı Renkler: Kırmızı, Yeşil, Siyah, Sarı


>>**Ordinal:** Anlamlı bir sıraya sahip olması gereken değerler

Örnek: Diyelim ki, A, B, C, D gibi performanslarına göre farklı notlar içerebilecek birkaç öğrencinin not sayfaları var.

 Notlar: A, B, C, D

>>**Sürekli:**  Sonsuz sayıda değerlere sahip olabilir, float türündendir.

Örnek: Birkaç Öğrencinin ağırlığını bir sıra veya düzenli bir şekilde ölçmek, yani 50, 51, 52, 53

Ağırlık: 50, 51, 52, 53


>>**Ayrık:** Sonlu sayıda değer

Örnek: Bir Öğrencinin Birkaç Dersteki Notları: 65, 70, 75, 80, 90 

Notları: 65, 70, 75, 80, 90


#**Sözdizimi:** 

>Matematiksel Gösterim: Sınıflandırma, girdi özellik vektörü “X” alan bir fonksiyon oluşturmaya ve “Y” sonucunu tahmin etmeye dayanır (C kümesindeki nitel yanıt alma değerleri) 
 
>Bir Denetimli fonksiyon olan Sınıflandırıcı (veya model) kullanılır, uzman bilgisine dayalı olarak manuel olarak tasarlanabilir. Sınıf etiketlerini tahmin etmek için yapılmıştır (Örnek: Etiket – bazı olayların onayı için “Evet” veya “Hayır”).
Sınıflandırıcılar iki ana türe ayrılabilir: 

**Ayırt edici :** Çok temel bir sınıflandırıcıdır ve her veri satırı için sadece bir sınıf belirler. Sadece gözlemlenen verilere bağlı olarak modellemeye çalışır, dağılımlardan ziyade büyük ölçüde verilerin kalitesine bağlıdır. 

**Üretken :** Bireysel sınıfların dağılımını modeller ve modelin varsayımlarını ve dağılımlarını tahmin ederek perde arkasındaki verileri üreten modeli öğrenmeye çalışır. Görünmeyen verileri tahmin etmek için kullanılır. 

Örnek : Naive Bayes Sınıflandırıcısı 
Önceki verilere bakarak Spam e-postaları algılama. 100 e-posta ve bunun da 1:4'e bölündüğünü varsayalım, yani A Sınıfı: %25 (Spam e-postalar) ve Sınıf B: %75 (Spam Olmayan e-postalar). Şimdi bir kullanıcı, bir e-postanın ucuz kelimesini içerip içermediğini kontrol etmek isterse, o zaman buna Spam denilebilir. 
Görünüşe göre A Sınıfında (yani verilerin %25'inde), 25 e-postadan 20'si spam ve geri kalanı değil. 
Ve B Sınıfında (yani verilerin %75'inde), 75 e-postadan 70'i spam değildir ve geri kalanı spamdır. 
Peki, e-posta ucuz kelimesini içeriyorsa, spam olma olasılığı nedir? (= %80)

**Dezavantajları:** 

Gizlilik: Veriler, bir şirketin müşterileri hakkında bazı bilgileri diğer satıcılara verme veya bu bilgileri kârları için kullanma şansı olduğunda. 
Doğruluk Problemi: En iyi doğruluğu ve sonucu alabilmek için Doğru model seçiminin olması gerekir. 


**VERİ MADENCİLİĞİNİN ÖZÜ:**

1.	Karar ağaçları, Bayes ağları veya sinir ağları gibi doğru sınıflandırma yöntemini seçme. 

2.	Tüm sınıf değerlerinin bilindiği bir veri örneğine ihtiyacınız var. Daha sonra veriler, eğitim seti ve test seti olmak üzere iki bölüme ayrılacaktır.
Şimdi eğitim seti, bir sınıflandırıcı türeten bir öğrenme algoritmasına verilir. Daha sonra sınıflandırıcı, tüm sınıf değerlerinin gizlendiği test seti ile test edilir. 

Sınıflandırıcı test kümesindeki çoğu durumu doğru sınıflandırırsa, gelecekteki veriler üzerinde de doğru çalıştığı varsayılabilir, aksi takdirde yanlış bir model seçilmiş olabilir.
