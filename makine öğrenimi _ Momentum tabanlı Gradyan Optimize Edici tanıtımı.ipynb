Gradient Descent , farklı modelleri eğitmek için Makine Öğrenimi çerçevelerinde kullanılan bir optimizasyon tekniğidir. Eğitim süreci, belirli bir veri kümesinde bir Makine Öğrenimi modelinin sahip olduğu hatayı belirleyen bir amaç işlevinden (veya hata işlevinden) oluşur. 
Eğitim sırasında bu algoritmanın parametreleri rastgele değerlere başlatılır. Algoritma yineledikçe, parametreler, fonksiyonun optimal değerine daha da yakınlaşacak şekilde güncellenir.
Bununla birlikte, Uyarlanabilir Optimizasyon Algoritmaları , hızla yakınsama yetenekleri nedeniyle popülerlik kazanmaktadır. Tüm bu algoritmalar, geleneksel Gradient Descent'in aksine, yakınsama sürecini sağlamlaştırmak için önceki yinelemelerdeki istatistikleri kullanır.
 

Momentum Tabanlı Optimizasyon:
Bir Uyarlamalı Optimizasyon Algoritması, yakınsamayı stabilize etmek için önceki yinelemelere göre gradyanların üstel ağırlıklı ortalamalarını kullanır ve bu da daha hızlı optimizasyon sağlar. Örneğin, Derin Sinir Ağlarının çoğu gerçek dünya uygulamasında, eğitim gürültülü veriler üzerinde gerçekleştirilir. Bu nedenle, Optimizasyon sırasında veriler yığınlar halinde beslendiğinde gürültünün etkisini azaltmak gereklidir. Bu sorun, Üstel Ağırlıklı Ortalamalar (veya Üstel Ağırlıklı Hareketli Ortalamalar) kullanılarak çözülebilir
