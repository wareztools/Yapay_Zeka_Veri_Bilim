Normal Denklem , En Küçük Kare Maliyet Fonksiyonlu Doğrusal Regresyona analitik bir yaklaşımdır. Gradient Descent kullanmadan doğrudan θ değerini bulabiliriz . Bu yaklaşımı izlemek, küçük özelliklere sahip bir veri kümesiyle çalışırken etkili ve zaman kazandıran bir seçenektir. Normal Denklem yöntemi, herhangi bir fonksiyonun türevinin ve kısmi türevinin minimum ve maksimum noktasında sıfır olacağı matematiksel Maxima & Minima kavramına dayanır. Böylece, Normal Denklem yönteminde, Maliyet fonksiyonunun her bir ağırlığa göre kısmi türevini bularak ve sıfıra eşitleyerek minimum değerini elde ederiz.
