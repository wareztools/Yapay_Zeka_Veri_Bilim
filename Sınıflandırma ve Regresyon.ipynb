##Sınıflandırma
Verileri birden çok kategorik sınıfa, yani ayrık değerlere ayırmaya yardımcı olan bir model veya işlevi bulma veya keşfetme sürecidir. Sınıflandırmada veriler, girişte verilen bazı parametrelere göre farklı etiketler altında sınıflandırılır ve ardından veriler için etiketler tahmin edilir. 
Türetilmiş eşleme işlevi “IF-THEN” kuralları şeklinde gösterilebilir. Sınıflandırma işlemi, verilerin ikili veya çoklu ayrık etiketlere bölünebildiği problemlerle ilgilenir. 

Bir örnek alalım, daha önce kaydedilen bazı parametrelere dayanarak A Takımının bir maçı kazanma olasılığını tahmin etmek istediğimizi varsayalım. O zaman Evet ve Hayır olmak üzere iki etiket olacaktır.

#Regresyon


Regresyon , sınıfları veya ayrık değerleri kullanmak yerine verileri sürekli gerçek değerlere ayırmak için bir model veya işlev bulma işlemidir. Ayrıca geçmiş verilere bağlı olarak dağıtım hareketini de tanımlayabilir. Bir regresyon tahmin modeli bir miktarı tahmin ettiğinden, modelin becerisi bu tahminlerde bir hata olarak rapor edilmelidir. 

Benzer bir örneği, daha önce kaydedilen bazı parametreler yardımıyla belirli bölgelerde yağmur olasılığını bulduğumuz regresyonda da ele alalım. O zaman yağmurla ilgili bir olasılık var.
