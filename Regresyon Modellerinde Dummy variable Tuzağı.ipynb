Regresyon Modellerinde Dummy Değişken: 
İstatistikte, özellikle regresyon modellerinde, çeşitli verilerle ilgileniriz. Veriler nicel (sayısal) veya nitel (kategorik) olabilir. Regresyon modellerinde sayısal veriler kolayca işlenebilir ancak kategorik verileri doğrudan kullanamayız, bir şekilde dönüştürülmesi gerekir. 

Kategorik öznitelikleri sayısal özniteliklere dönüştürmek için etiket kodlama prosedürünü kullanabiliriz (etiket kodlaması her veri kategorisine benzersiz bir tamsayı atar). Ancak bu prosedür tek başına uygun değildir, bu nedenle regresyon modellerinde etiket kodlamasını takiben one-hot kodlama kullanılır. Bu, kategorik öznitelikte bulunan sınıfların sayısına göre yeni öznitelikler oluşturmamızı sağlar, yani kategorik öznitelikte n sayıda kategori varsa, n yeni öznitelik oluşturulacaktır. Oluşturulan bu niteliklere dummy Değişkenler denir . Bu nedenle, dummy değişkenler, regresyon modellerinde kategorik veriler için “vekil” değişkenlerdir. 
Bu dummy değişkenler,one-hot kodlama ile oluşturulacaktır.ve her öznitelik, o özniteliğin varlığını ya da yokluğunu temsil eden 0 ya da 1 değerine sahip olacaktır. 

##Dummy Değişken Tuzağı: 
Dummy değişken tuzağı, yüksek düzeyde ilişkili (Çok Yönlü) niteliklerin olduğu ve bir değişkenin diğerlerinin değerini tahmin ettiği bir senaryodur. Kategorik verileri işlemek için one-hot kodlamayı kullandığımızda , diğer dummy değişkenlerin yardımıyla bir dummy değişken (nitelik) tahmin edilebilir. Bu nedenle, bir dummy değişken, diğer dummy değişkenlerle yüksek oranda ilişkilidir. Regresyon modelleri için tüm dummy değişkenlerin kullanılması dummy değişken tuzağına yol açar . Bu nedenle, regresyon modelleri bir dummy değişkeni hariç tutacak şekilde tasarlanmalıdır.
