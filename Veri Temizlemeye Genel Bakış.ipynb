#Giriş
Veri temizleme, makine öğreniminin önemli parçalarından biridir. Bir model oluşturmada önemli bir rol oynar. Bu kesinlikle makine öğreniminin en zor kısmı değildir ve  ortaya çıkarılacak gizli numaralar veya sırlar da yoktur. Ancak, bir projenin başarısı veya başarısızlığı, uygun veri temizliğine bağlıdır. Profesyonel veri bilimcileri, “Daha iyi veri, daha meraklı algoritmaları yener” inancı nedeniyle genellikle zamanlarının çok büyük bir bölümünü bu adıma harcarlar . 


İyi temizlenmiş bir veri kümemiz varsa, basit algoritmalarla da iyi sonuçlar elde etme şansımız vardır, bu da veri kümesi boyutunun büyük olduğu durumlarda özellikle hesaplama açısından zaman zaman çok faydalı olabilir. 

Açıkçası, farklı veri türleri farklı temizlik türleri gerektirecektir. Ancak, bu sistematik yaklaşım her zaman iyi bir başlangıç ​​noktası olarak hizmet edebilir.

##İstenmeyen Gözlemlerin Kaldırılması 
Veri kümenizden yinelenen/yedek veya alakasız değerlerin silinmesi dahildir. 

Yinelenen ve alakasız gözlemler en sık veri toplama sırasında ortaya çıkar ve  çözmeye çalıştığınız belirli soruna uymayan gözlemlerdir. 

Gereksiz gözlemler, veriler tekrarlandıkça verimliliği büyük ölçüde değiştirir ve doğru tarafa veya yanlış tarafa doğru ekleme yapabilir, bu nedenle aldatıcı sonuçlar üretebilir.

Alakasız gözlemler, bize faydası olmayan ve doğrudan kaldırılabilen her türlü veridir.

##Yapısal Hataların Düzeltilmesi
 Ölçüm, veri aktarımı veya benzeri durumlarda ortaya çıkan hatalara yapısal hatalar denir. 
 
Yapısal hatalar, özelliklerin adındaki yazım hatalarını, farklı bir adla aynı özniteliği, yanlış etiketlenmiş sınıfları, yani gerçekten aynı olması gereken ayrı sınıfları veya tutarsız büyük harf kullanımını içerir.

Örneğin, bir sınıf diğer iki sınıfa dahil edilebilse de, model Türkiye ve türkiye'yi farklı sınıflar veya değerler olarak ele alacaktır, ancak aynı değeri veya kırmızı, sarı ve kırmızı-sarı'yı ​​farklı sınıflar veya nitelikler olarak temsil ederler. Dolayısıyla bunlar, modelimizi verimsiz yapan ve kalitesiz sonuçlar veren bazı yapısal hatalardır. 

##İstenmeyen Aykırı Değerleri Yönetme
Aykırı Değerler, belirli model türlerinde sorunlara neden olabilir.

 Örneğin, doğrusal regresyon modelleri, aykırı değerlere karşı karar ağacı modellerinden daha az dayanıklıdır. 
 
 Genel olarak, bunları kaldırmak için meşru bir nedenimiz olana kadar aykırı değerleri kaldırmamalıyız. Bazen bunları kaldırmak performansı artırır, bazen de artırmaz. 
 
 Bu nedenle, gerçek verilerin parçası olma olasılığı düşük olan şüpheli ölçümler gibi aykırı değerleri ortadan kaldırmak için iyi bir nedeniniz olmalıdır.

##Eksik Verileri İşleme
Eksik veriler, makine öğreniminde aldatıcı bir şekilde yanıltıcı bir sorundur.

 Eksik gözlemi görmezden gelemeyiz veya kaldıramayız. 
 
 Önemli bir şeyin göstergesi olabileceğinden dikkatli bir şekilde ele alınmaları gerekir. 
 
 Eksik verilerle başa çıkmanın en yaygın iki yolu şunlardır: 
> **Eksik değerlerle gözlemler bırakılıyor.**
Değerin eksik olması başlı başına bilgilendirici olabilir.
Ayrıca, gerçek dünyada, bazı özellikler eksik olsa bile, genellikle yeni veriler üzerinde tahminler yapmanız gerekir.

> **Geçmiş gözlemlerden eksik değerleri uygulamak.**
“eksiklik” neredeyse her zaman kendi içinde bilgilendiricidir ve bir değer eksikse algoritmada belirtilmelidir.
Değerlerinizi empoze etmek için bir model oluştursanız bile, gerçek bir bilgi eklemiyorsunuz. Diğer özellikler tarafından zaten sağlanan kalıpları güçlendiriyorsunuz.

Eksik veri, bir yapboz parçasını kaçırmak gibidir. Eğer düşürürseniz, bu yapboz yuvası yokmuş gibi davranmak gibidir. Eğer onu uydurmaya çalışırsanız , bu yapbozun başka bir yerinden bir parçayı sıkıştırmaya çalışmak gibidir. 
Bu nedenle, eksik veriler her zaman bilgilendirici ve önemli bir şeyin göstergesidir.

Eksik verileri işaretleyerek algoritmamızın farkında olmalıyız. Bu işaretleme ve doldurma tekniğini kullanarak, algoritmanın sadece ortalama ile doldurmak yerine, eksiklik için en uygun sabiti tahmin etmesine izin veriyorsunuz.
